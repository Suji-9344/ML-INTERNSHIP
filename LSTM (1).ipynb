```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
import math
from sklearn.metrics import mean_squared_error
```

```python
data= pd.read_csv('/content/drive/MyDrive/IBM_2006-01-01_to_2018-01-01 (1).csv',index_col='Date',parse_dates=['Date'])
data.head()
```

```python
mytrain = data.loc[:'2016', data.columns[1:2]].values
mytest = data.loc['2017':, data.columns[1:2]].values
print(mytrain[0])
print(mytest[0])
```

```python
sc = MinMaxScaler(feature_range=(0,1))
mytrain_scaled=sc.fit_transform(mytrain)
print(mytrain_scaled)
```

```python
len(mytrain_scaled)
```

```python
I_train = []
o_train = []
for i in range(60,2769):
  I_train.append(mytrain_scaled[i-60:i,0])
  o_train.append(mytrain_scaled[i,0])
```

```python
I_train = np.array(I_train)
o_train = np.array(o_train)
print(I_train)
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(I_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1)
```

```python
model.compile(optimizer= 'adam',loss='mean_squared_error')
```

```python
model.fit(I_train,o_train,epochs=50,batch_size=32)
```

```python
model.save('/content/drive/MyDrive/stock_price.h5')
```
