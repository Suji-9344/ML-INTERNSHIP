```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/random.csv.zip")
df.head()
```

```python
transactions = []

for i in range(len(df)):
    transactions.append([str(df.iloc[i, 0])])
    print(transactions)
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df_encoded = pd.DataFrame(te_array, columns=te.columns_)
df_encoded.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_items = apriori(
    df_encoded,
    min_support=0.2,
    use_colnames=True
)

frequent_items
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Create encoder instance
te = TransactionEncoder()

# Fit & transform
te_array = te.fit(transactions).transform(transactions)

# Create encoded DataFrame
encoded_df = pd.DataFrame(te_array, columns=te.columns_)

encoded_df.head()
```

```python
import matplotlib.pyplot as plt

if not frequent_items.empty:
    plt.figure()
    plt.bar(frequent_items['items'], frequent_items['support'])
    plt.xlabel("Itemsets")
    plt.ylabel("Support")
    plt.title("Frequent Itemsets Support")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No frequent itemsets found to plot. The 'frequent_items' DataFrame is empty.")
    print("Please ensure the 'apriori' algorithm generated frequent itemsets with the chosen min_support.")
```
