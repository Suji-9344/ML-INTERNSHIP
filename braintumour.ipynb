```python

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
IMG_SIZE=224
BATCH_SIZE=32
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,validation_split=0.2)
```

```python
train_generator=train_datagen.flow_from_directory(
'/content/drive/MyDrive/brain_tumour_dataset/train',
target_size=(IMG_SIZE,IMG_SIZE),
batch_size=BATCH_SIZE,
class_mode='binary',
subset='training'
)
```

```python
val_generator=train_datagen.flow_from_directory(
target_size=(IMG_SIZE,IMG_SIZE),
batch_size=BATCH_SIZE,
class_mode='binary',
subset='validation'
)
```

```python
model=keras.sequential([
    layers.conv2D(32,(3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    layers.maxpooling2D((2,2)),
    layers.con2D(64,(3,3), activation='relu'),
    layers.maxpooling((2,2)),
    layers.con2D(128,(3,3), activation='relu'),
    layers.maxpooling((2,2)),
    layers.flattern(),
    layers.dense(128, activation='relu'),
    layers.dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimize='adam',loss='binary_crossentropy',matrices=['accuracy'])
```

```python
model.fit(train_generator,epochs=5,validation_dataset=val_generator.batch_size=BATCH_SIZE)
```

```python
model.save()
```
